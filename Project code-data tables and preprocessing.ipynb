```python
# this is the code for extracting the date and cumulatively taking the tweets for the entire day and taking the sum of its followers in that day

datadict={}
followerdict={}
datelist=df1.Date.unique()
count=0
temp=[]
follower_count=0
for i in range(len(df1.index)):
    if datelist[count]==df1.iloc[i]['Date']:
        temp.append(df1.iloc[i]['Tweet content'])
        follower_count+=df1.iloc[i]['Followers']
    else:
        datadict[datelist[count]]=temp
        followerdict[datelist[count]]=follower_count
        count+=1
        temp=[]
        follower_count=0
        temp.append(df1.iloc[i]['Tweet content'])
        follower_count+=df1.iloc[i]['Followers']   
```
