```python
from pandas_datareader import data as web
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
```

```python
start=datetime.datetime(2016, 3, 30)
end=datetime.datetime(2016, 6, 15)
print(start,end)

df = web.DataReader(['AAPL'], 'yahoo', start, end)
df.reset_index(level=0, inplace=True)
df.plot(x='Date', y= 'Close',figsize=(10, 6),color="r")  
plt.show()
# rets = df['Open'].pct_change()
# df['rets'] = np.where(df['Open'].pct_change())
# rets
```
